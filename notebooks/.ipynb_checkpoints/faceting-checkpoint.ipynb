* Create a 4 row and 1 column plot where:
    * The first row is a line graph of all the first quater earnings
    * The second row is a line graph of all the second quater earnings
    * The third row is a line graph of all the third quater earnings
    * The fourth row is a line graph of all the fourth quater earnings
* All the plots should have a title saying what quarter is being shown
* All the plots plots but the final should not have an x-axis, the final plot should only have the even years on the axis
* The y-axis should be called "Earnings"
* The x-axis should be called "Year"
* The whole plot should be titled "Johnson and Johnson Quarterly Earnings"


````python
# Your code below (You need to make the appropriate imports of matplotlib and numpy)
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
col1=pd.Series([0.71, 0.61,0.72, 0.83, 0.92, 1.16,1.26,1.53,1.53,2.16, 2.79, 3.60,4.86, 5.58,6.03,6.93,7.74,9.54,11.88,14.04,16.20])
col2=pd.Series([0.63, 0.69,0.77, 0.80, 1.00, 1.30,1.38,1.59,2.07,2.43, 3.42, 4.32,5.04, 5.85,6.39,7.74,8.91,10.26,12.06,12.96,14.67])
col3=pd.Series([0.85, 0.92,0.92, 1.00, 1.24, 1.45,1.86,1.83,2.34,2.70, 3.69, 4.32,5.04, 6.57,6.93,7.83,8.28,9.54,12.15,14.85,16.02])
col4=pd.Series([0.44, 0.55,0.60, 0.77, 1.00, 1.25,1.56,1.86,2.25,2.25, 3.60, 4.05,4.41, 5.31,5.85,6.12,6.84,8.73,8.91,9.99,11.61])
stamps=pd.Series(np.linspace(1960,1981,21,dtype = int,endpoint=False))
stamps
johnson_df=pd.DataFrame({'time':stamps,
                         'Qtr1':col1, 
                         'Qtr2':col2,
                          'Qtr3':col3,
                          'Qtr4':col4
                           })
johnson_df

plt.figure()
plt.figtext(0.5, -0.1, "Johnson and Johnson Quarterly Earnings", ha="center", fontsize=12)

# # Creating first subplot
plt.subplot(4, 1, 1)
plt.plot(johnson_df['time'],johnson_df['Qtr1'],label='Quarter 1')
plt.xticks([])
plt.ylabel('Earnings')
plt.title('Q1')

# # Creating second subplot
plt.subplot(4, 1, 2)
plt.plot(johnson_df['time'],johnson_df['Qtr2'],label='Quarter 2')
plt.xticks([])
plt.ylabel('Earnings')
plt.title('Q2')

# # Creating third subplot
plt.subplot(4, 1, 3)

plt.plot(johnson_df['time'],johnson_df['Qtr3'],label='Quarter 3')
plt.ylabel('Earnings')
plt.xticks([])
plt.title('Q3')



# # Creating fourth subplot
plt.subplot(4, 1, 4)

plt.plot(johnson_df['time'],johnson_df['Qtr4'],label='Quarter 4')
plt.xlabel('year')
plt.ylabel('Earnings')
tick_location =np.arange(1960,1981,2)
tick_labels =np.arange(1960,1981,2)
plt.xticks(tick_location, tick_labels)
plt.title('Q4')


plt.tight_layout()
plt.show()
````

### Challenge 2

For this challenge you will be working with the **cars** data set. Run the cell below to import the data set.

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

cars_data = pd.read_csv('../data/cars.csv', sep='|')
cars_data.rename(columns={'Unnamed: 0': 'make_model'}, inplace=True)
cars_data.head()
````

For this challenge:
* Facet the plots on the number of cylenders each vechical has
* Create a single row with with the number of correct plots
* Use a scatter plot to show the relation between horse power and quarter mile time
* On the x axis plot the horse power
* On the y axis plot the the quarter mile time (qsec)
* Each plot should have a title saying the number of cylinders
* Give each axis an appropriate name

````python
# Write your code here
num_cylinders=len(cars_data.cyl.unique())
num_cylinders

# We need 3 subplots for data visualization
plt.figure()
plt.figtext(0.5, -0.1, "Scatterplot of Quarter mile time vs Horse Power", ha="center", fontsize=12)

# # Creating first subplot

plt.subplot(1, 3, 1)
plt.scatter(cars_data.loc[cars_data['cyl']==4,:]['hp'],cars_data.loc[cars_data['cyl']==4,:]['qsec'], marker='o',color='blue')
plt.xlabel('Horse Power')
plt.ylabel('Quarter mile time')
plt.title('Cylinder=4')

# # Creating second subplot


plt.subplot(1, 3, 2)
plt.scatter(cars_data.loc[cars_data['cyl']==6,:]['hp'],cars_data.loc[cars_data['cyl']==6,:]['qsec'], marker='o',color='blue')
plt.xlabel('Horse Power')
plt.yticks([])
plt.title('Cylinder=6')

# # Creating third subplot



plt.subplot(1, 3, 3)
plt.scatter(cars_data.loc[cars_data['cyl']==8,:]['hp'],cars_data.loc[cars_data['cyl']==8,:]['qsec'], marker='o',color='blue')
plt.xlabel('Horse Power')
plt.yticks([])
plt.title('Cylinder=8')

plt.tight_layout()
plt.show()
````
